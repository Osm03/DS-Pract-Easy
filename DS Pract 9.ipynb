```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')
```

```python
sns.barplot(x='sex',y='age',data=titanic,palette='husl')
```

```python
# Set the style of the plot using Seaborn
sns.set_style('darkgrid')

# Create a figure and axis object using Matplotlib
fig, ax = plt.subplots(figsize=(12,6))

# Use Seaborn's boxplot to visualize the distribution of ages for each gender and survival status
sns.boxplot(data=titanic, x='sex', y='age', hue='survived', palette='husl', ax=ax)

# Add a title and axis labels using Matplotlib
ax.set_title('Distribution of Age by Gender and Survival Status', fontsize=16)
ax.set_xlabel('Gender', fontsize=14)
ax.set_ylabel('Age', fontsize=14)

# Add a legend to clarify the meaning of 0 and 1 in the 'survived' variable
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Did Not Survive', 'Survived'])


# Show the plot using Matplotlib
plt.show()
```
