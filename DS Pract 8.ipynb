```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
```

```python
sns.countplot(x='class', data=df)
```

```python
sns.catplot(x ="sex", hue ="survived",kind ="count", data = df)
```

```python
# Create a scatterplot of age vs. fare, colored by class
sns.scatterplot(x='age', y='fare', hue='class', data=df)
```

```python
# Create a histogram of ages, with separate colors for men and women
sns.histplot(data=df, x='age', hue='sex')
```

```python
# Create a countplot of passenger survival, colored by class
sns.countplot(x='survived', hue='class', data=df)
```

```python
# Create a figure and axis object using Matplotlib
fig, ax = plt.subplots(figsize=(8,6))

# Use Seaborn's histplot to visualize the distribution of fares
sns.histplot(data=df, x='fare', kde=True, stat='density', color='teal', alpha=0.7, ax=ax)

# Add a title and axis labels using Matplotlib
ax.set_title('Distribution of Ticket Fares on the Titanic', fontsize=16)
ax.set_xlabel('Ticket Fare', fontsize=14)
ax.set_ylabel('Density', fontsize=14)

# Show the plot using Matplotlib
plt.show()
```

```python
# Create a figure and axis object using Matplotlib
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))

# Use Seaborn's histplot to visualize the distribution of fares for male and female passengers
sns.histplot(data=df[df['sex'] == 'male'], x='fare', kde=True, stat='density', color='teal', alpha=0.7, ax=ax1)
sns.histplot(data=df[df['sex'] == 'female'], x='fare', kde=True, stat='density', color='purple', alpha=0.7, ax=ax2)

# Add a title and axis labels using Matplotlib
ax1.set_title('Distribution of Ticket Fares for Male Passengers', fontsize=16)
ax1.set_xlabel('Ticket Fare', fontsize=14)
ax1.set_ylabel('Density', fontsize=14)

ax2.set_title('Distribution of Ticket Fares for Female Passengers', fontsize=16)
ax2.set_xlabel('Ticket Fare', fontsize=14)
ax2.set_ylabel('Density', fontsize=14)

# Show the plot using Matplotlib
plt.show()
```
